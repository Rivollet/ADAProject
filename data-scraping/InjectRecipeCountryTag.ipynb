## Metadata Functions

```python
def importMD(path):
    
    metadata = []
    # Import existing data
    try:
        with open(path, 'r') as file:
            metadata = json.load(file)
            file.close()
    except IOError as e:
        print('There is no meta-data to import, a default data structure is created')
    return metadata


def isCategoryExist(metadata, id):
    
    # Find the category
    category = None
    for (i, cat) in enumerate(metadata):
        if(cat['id'] == id):
            return True
    return False

def isPresentMD(metadata, id, page):
    
    # Find the category
    category = None
    for (i, cat) in enumerate(metadata):
        if(cat['id'] == id):
            category = cat
            break;
    
    if (category != None):
        return (page in category['pages'])
    else:
        return False
```

## Functions

```python
def splitUrlInStrings(url):
    return url.split('/')[3:-1]

def importTags(path):
    data = []
    with open(path, 'r') as file:
            data = json.load(file)
    return data

def exportTags(data, path):
    with open(path, 'w') as file:
        json.dump(data, file, sort_keys=True, indent=4)
```

## Mongo Function

```python
def FindIdFromRecipeId(recipeId):
    out = collection.find_one({'recipeID':recipeId})
    return out['_id']

def AddTagsToRecipe(recipeId, values):
    _id = FindIdFromRecipeId(recipeId)
    collection.update_one({'_id':_id}, {'$addToSet': {'ada-tags': {'$each':values} }})
```
